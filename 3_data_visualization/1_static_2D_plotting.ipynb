# Static 2D plotting

## One-Dimensional Data Sets

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mpl.__version__
```

```python
sns.set(style="whitegrid") 
mpl.rcParams['font.family'] = 'serif'
%matplotlib inline
```

```python
# x,y shows the lists of arrays containing the values of the abscissa and ordinate respectively
np.random.seed(1000)
```

```python
y = np.random.standard_normal(20)
```

### Plot the given x and y values

```python
x = np.arange(len(y))
plt.plot(x, y)
```

### Plot given data as an ndarray object

```python
plt.plot(y)
# if we only provide the y values, plt.plot() takes the index values at the respective x values.
```

### Cumulative Sum

```python
plt.plot(y.cumsum())
```

### Plot without grid
Different plotting styles may be required for different publications and different requirements

```python
plt.plot(y.cumsum())
plt.grid(False)
plt.axis('equal')
```

### Plot with Custom Axis Limits

```python
plt.plot(y.cumsum())
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) -1,
         np.max(y.cumsum()) +1)
```

### Plot with Typical Labels

```python
plt.figure(figsize=(10, 6))
plt.plot(y.cumsum(), 'b', lw = 1.5)
plt.plot(y.cumsum(), 'ro')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('A simple plot with typical labels')
```

## Two Dimensional Datasets

### Plot with two datasets

```python
# generates a two-dimensional sample data set as NumPy ndarray object of shape 20 X 2 with standard normally distributed pseudo-random numbers.
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y, lw =1.5)
plt.plot(y, 'ro')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title(' A Plot with two datasets')
```

### Plot with labeled data sets

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw = 1.5, label = '1st')
plt.plot(y[:, 1], lw = 1.5, label = '2nd')
plt.plot(y, 'ro')
plt.legend(loc = 0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title(' A simple plot with labeled data sets')
```

### Plot with two differently scaled datasets

```python
# it scales the first two data subsets by a factor of 100 and plots the data again
y [:, 0] = y[:, 0 ] * 100
```

```python
plt.figure(figsize=(10, 6))
plt.plot(y[:, 0], lw = 1.5, label = '1st')
plt.plot(y[:, 1], lw = 1.5, label = '2nd')
plt.plot(y, 'ro')
plt.legend(loc = 0)
plt.xlabel('Index')
plt.ylabel('Value')
plt.title(' A simple plot with two differently scaled data sets')
```

From the inspection of the above plot, we can see that the first data is still "visually readable", while the second data set looks like a straight line with the new scaling

### Plot with two data sets and two y - axis

```python
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw = 1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 8)
plt.xlabel('Index')
plt.ylabel('Value 1st')
plt.title('A Simple Plot with two data sets and two y-axis')
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw = 1.5, label = '2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc = 0)
plt.ylabel("Value 2nd")
```

## Plot with two sub plots

```python
plt.figure(figsize=(10, 6))
plt.subplot(211)
plt.plot(y[:, 0], lw = 1.5, label = '1st')
plt.plot(y[:, 0], 'ro')
plt.legend(loc = 0 )
plt.ylabel('Value')
plt.title('Plot with two sub plots')
plt.subplot(212)
plt.plot(y[:, 1], lw = 1.5, label = '2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc = 0)
plt.xlabel('Index')
plt.ylabel('Value')
```
